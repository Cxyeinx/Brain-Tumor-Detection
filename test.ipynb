```python
import tensorflow as tf
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = tf.keras.models.load_model("model.model")
```

```python
normal_img = Image.open("normal.png").convert("RGB").resize((128,128))
normal_img = np.array(normal_img)
normal_img = np.expand_dims(normal_img, axis=0)
normal_img = normal_img / 255
```

```python
effected_img = Image.open("effected.png").convert("RGB").resize((128,128))
effected_img = np.array(effected_img)
effected_img = np.expand_dims(effected_img, axis=0)
effected_img = effected_img / 255
```

```python
print(np.argmax(model.predict(normal_img)))
plt.imshow(normal_img[0])
```

```python
print(np.argmax(model.predict(effected_img)))
plt.imshow(effected_img[0])
```
